1. Viết function thực hiện đánh giá classification model bằng F1-Score.

```python
# Input: tp, fn, fn.
# Output: precision, recall, F1-Score.
import sys

def exercise1(tp,fp,fn): 
    # Check int
    parameters_lst = [tp,fp,fn]
    name_parameters = ['tp','fp','fn']
    for name,num in zip(name_parameters,parameters_lst):
        if not isinstance(num, int):
            print(f'{name} must be int')
            sys.exit(0) # Stop

        if num <= 0:
            print('tp and fp and fn must be greater than zero')
            sys.exit(0) # Stop

    # Formulars
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1_score = 2 * (precision * recall) / (precision + recall)

    # Print them
    print(f'precision is {precision}')
    print(f'recall is {recall}')
    print(f'f1-score is {f1_score}')

# Run
exercise1(tp=3, fp=2, fn=4)
```

2. Viết function mô phỏng theo 3 activation fuction

```python
""" 
Input: 
- Người dùng nhập giá trị x
- Người dùng nhập tên activation chỉ có 3 loại (sigmoid, relu, elu)

Output:
Kết quả f(x)
"""

import math
import sys

def exercise2():
    
    def is_number(n):
        try:
            float(n)    # Type-casting the string to 'float'
                        # If string is not a valid 'float',
                        # it'll raise 'ValueError' exception.
        
        except ValueError:
            return False
        return True,n
    def is_valid_function(function):
        available_function = ['sigmoid','relu','elu']
        if function in available_function:
            return True
        else:
             return False
         
    x = input('Input x = ')
    input_function = input('Input activation function (sigmoid|relu|elu): ')
    
    if not is_number(x):
        print('x must be a number')
        sys.exit(0)
        
    x = float(x)

    if not is_valid_function(input_function):
        print(f'{input_function} is not supported')
        sys.exit(0)
    
    alpha = 0.01
    
    #Sigmoid formular
    sigmoid = 1 / (1 + math.e**(-x))
    
    # Relu formular
    if x <= 0:
        relu = 0
    else:
        relu = x
    
    # Elu formular
    if x <= 0:
        elu = alpha*(math.e**x -1)
    else:
        elu = x
    
    # Display
    if input_function == 'sigmoid':
        print(f'{input_function}: f({x}) = {sigmoid}')
    elif input_function == 'relu':
        print(f'{input_function}: f({x}) = {relu}')
    else:
        print(f'{input_function}: f({x}) = {elu}')

exercise2()
```

3. Viết function lựa chọn regression los function để tính loss

```python
"""
Input:
- Người dùng nhập số lượng sample (num_samples) được tạo ra (chỉ nhận integer
numbers)
- Người dùng nhập loss name (MAE, MSE, RMSE-(optional)) chỉ cần MAE và
MSE, bạn nào muốn làm thêm RMSE đều được.
Output: Print ra loss name, sample, predict, target, loss
- loss name: là loss mà người dùng chọn
- sample: là thứ tự sample được tạo ra (ví dụ num_samples=5, thì sẽ có 5 samples và
mỗi sample là sample-0, sample-1, sample-2, sample-3, sample-4)
- predict: là số mà model dự đoán (chỉ cần dùng random tạo random một số trong range
[0,10))
- target: là số target mà momg muốn mode dự đoán đúng (chỉ cần dùng random tạo
random một số trong range [0,10))
- loss: là kết quả khi đưa predict và target vào hàm loss
- note: ví dụ num_sample=5 thì sẽ có 5 cặp predict và target.
"""
```

```python
import sys
import random

def exercise3():
    num_sample = input('Input number of sampels (integer number) which are generated: ')
    loss_name = input('Input loss name: ')
    # Checking
    if not num_sample.isnumeric():
        print('number of samples must be an integer number')
        sys.exit(0)
        
    num_sample = int(num_sample)    # Type-casting to int
     
    def abs_y1y2(y1,y2):    # y1: target, y2: predict
        return math.abs(y1-y2)
    
    def quadratic_y1y2(y1,y2):  # y1: target, y2: predict
        return (y1-y2)**2
    
    def mae_loss(n,sum):
        return sum / n
    
    def mse_loss(n,sum):
        return sum / n
    
    def rmse_loss(n,sum):
        return math.sqrt(sum / n)
    
    sum_loss = 0
    for i in range(5):
        # Generate
        pred = random.uniform(0,10)
        target = random.uniform(0,10)
        
        # Compute loss value
        if loss_name == 'MAE':
            loss = abs_y1y2(pred,target)
        elif loss_name == 'MSE' or loss_name == 'RMSE':
            loss = quadratic_y1y2(pred,target)
        
        sum_loss += loss    # Total losses
        # Display results
        print(f'''loss name: {loss_name}, sample: {num_sample}, pred: {pred}, target: {target},
        loss: {loss}''')
    
    # Display final loss
    if loss_name == 'MAE':
        final_value = mae_loss(num_sample,sum_loss)
    elif loss_name == 'MSE':
        final_value = mse_loss(num_sample,sum_loss)
    elif loss_name == 'RMSE':
        final_value = rmse_loss(num_sample,sum_loss)
    print(f'Final {loss_name}: {final_value}')

exercise3()
        
    
```

4. Viết 4 functions để ước lượng các hàm số sau

```python
"""
Input: x (số muốn tính toán) và n (số lần lặp muốn xấp xỉ)
Output: Kết quả ước lượng hàm tương x
"""
def factorial(n):
    if n == 1 or n==0:
        return 1
    else:
        return n * factorial(n-1)

def approx_sin(x,n):
    sin_sum = 0
    for e in range(n):
        sin_sum += ((-1)**e) * x**(2*e + 1) / factorial(2*e +1)
    return sin_sum

def approx_cos(x,n):
    cos_sum = 0
    for e in range(n):
        cos_sum += ((-1)**e) * x**(2*e) / factorial(2*e)
    return cos_sum

def approx_sinh(x,n):
    sinh_sum = 0
    for e in range(n):
        sinh_sum += x**(2*e + 1) / factorial(2*e + 1)
    return sinh_sum

def approx_cosh(x,n):
    cosh_sum = 0
    for e in range(n):
        cosh_sum += x**(2*e) / factorial(2*e)
    return cosh_sum

approx_cosh(x=3.14, n=10)
# approx_cos(x=3.14, n=10)
```
